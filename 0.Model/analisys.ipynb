```python
count_list = []
for i in range(10):
    s1 = pd.Series(t.iloc[i,:]).tolist()   # 1열 ~ 10열 
    for j in range(len(s1)):
        count_list.append(s1[j])
    print(len(count_list), f'{j}개 추가 됨')
len(count_list)
```

```python
title_count = pd.Series(count_list).value_counts()[:100]
```

```python
view_data = []
num = []
for i, title in enumerate(title_count):
    print(f"{i+1} 위", end=': '), print(title_count.index[i], title)
    line = (f'{i+1} 위: {title_count.index[i]}')
    view_data.append(line)
    num.append(title)
len(view_data), len(num)
```
