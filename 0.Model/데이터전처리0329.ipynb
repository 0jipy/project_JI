```python
temp1 = pd.read_csv(r'./save/0329result/temp1_wavve.csv')
temp2 = pd.read_csv(r'./save/0329result/temp2_wacha.csv')
temp3 = pd.read_csv(r'./save/0329result/temp3_netflix.csv')
temp4 = pd.read_csv(r'./save/0329result/temp4_disney.csv')
```

- 마킹 및 중복값 드랍

```python
# 마킹 컬럼 4개 추가 및 0, 1 마킹.
temp1 = temp1.assign(m_wavve = 1, m_wacha = 0, 
            m_netflix = 0, m_disney = 0)
temp2 = temp2.assign(m_wavve = 0, m_wacha = 1, 
            m_netflix = 0, m_disney = 0)
temp3 = temp3.assign(m_wavve = 0, m_wacha = 0, 
            m_netflix = 1, m_disney = 0)
temp4 = temp4.assign(m_wavve = 0, m_wacha = 0, 
            m_netflix = 0, m_disney = 1)
```

```python
len(temp1.columns)  # 9 + 4개 13개 
```

- 로우수 적은 디즈니부터 공통된 영화 찾고 & 마킹 >>> 컨캣후 중복값 드랍. 

```python
# 3과 4  (넷플과 디즈니)  

t1 = pd.merge(temp3, temp4, on=['title','opening_date'],how='inner')
t1.head()
```

```python
# 4개의 영화 공통. >>> 매핑 temp3 디즈니에 1 

# 인덱스로 값을 찾아 부여하기 위해 
length = range(len(t1))
for i in length:
    for j in range(len(temp3)):
        if (temp3.title.values[j] + str(temp3.opening_date.values[j]) == t1.title.values[i] + str(t1.opening_date.values[i])):
            temp3.m_disney[j] = 1  # temp3에 디즈니 매핑
            print(i), print(j)

print("=== "+ str(sum(temp3.m_disney[:]))+" 개의 공통 영화 매핑 완료 ===")
```

```python
# 컨캣 후 드랍 # 색인을 위해 일부러 인덱스는 남김
dt_concat = pd.concat([temp3, temp4], axis=0)
len(dt_concat) , len(temp3), len(temp4) # 3390 + 969 = 4359
```

```python
# 중복값 체크후 없으므로 드랍따로 없음
dt_concat.duplicated().sum()   # >>> 0
# 공통된 4개 제거된 숫자 , inplace=None 이라 데이터 변화 없음. 
len(dt_concat.drop_duplicates(subset=['title', 'opening_date']))
```
